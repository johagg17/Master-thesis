#### Compose a dataset with only admitted people

```python
parent_path = 'D:\skola√Ör5\MasterThesis\mimic-iv-1.0'
df_patients = pd.read_csv(parent_path + '\core\patients.csv')
df_admissions = pd.read_csv(parent_path + r'\core\admissions.csv')
```

```python
df_patadm = pd.merge(df_patients, df_admissions, on='subject_id')
```

To make things a bit easier, lets start by focusing on three patients, which have at least have 5 appointments.

```python
admperpat = df_patadm.groupby(['subject_id']).size().reset_index(name='count')
```

```python
more_than_5appointments = admperpat[admperpat['count'] > 5]
```

Randomly select 3 patients to investigate.

```python
more_than_5appointments['subject_id']
```

```python
patientsids = more_than_5appointments.sample(n=1)['subject_id'].tolist()
```

```python
patientsids
```

```python
df_final = df_patadm[df_patadm['subject_id'].isin(patientsids)].sort_values('admittime')
```

```python
df_final.head()
```

```python
df_final['1'] = 1
```

```python
df_final['admittime'] = pd.to_datetime(df_final['admittime'])
df_final['dischtime'] = pd.to_datetime(df_final['dischtime'])
age_firstadm = df_final['anchor_age'].iloc[0]
year_firstadm = df_final['admittime'].iloc[0].year
df_final['time_stay'] = abs(df_final['admittime'] - df_final['dischtime']).apply(lambda x: x.days)
df_final['size'] = abs(df_final['admittime'] - df_final['dischtime']).apply(lambda x: x.days*3 if x.days < 10 else int(x.days) * 0.5)
df_final['age'] = df_final['admittime'].apply(lambda x: abs(x.year - year_firstadm) + age_firstadm)
```

Plot scatter of patients and each admission

```python
def admissions(patient_id):
    
    
    p = figure(title = "Admissions for patient {} ".format(patient_id), plot_width=800, plot_height=350)
    p.xaxis.formatter=DatetimeTickFormatter(
        hours=["%d %B %Y"],
        days=["%d %B %Y"],
        months=["%d %B %Y"],
        years=["%d %B %Y"],
    )
    tooltips = [('Age', '@age'), 
                ('Admission Stay', '@time_stay'),
               ('Admission type', '@admission_type')]
    
    p.add_tools(HoverTool(tooltips=tooltips))
    
    source = ColumnDataSource(data=df_final)
    
    p.rect(x='admittime', y='1', width=5, height=20, color="#CAB2D6",
          width_units="screen", height_units="screen", source=source)
    
    return p
```
